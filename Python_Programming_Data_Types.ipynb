#1.6 Strings

In the world of programming, we call text a String.

Think of it like a "string of pearls," where each pearl is a character (a letter, number, or space) linked together in a specific order.

In Python, a string is a sequence of characters.You can create them by wrapping your text in:
- Single quotes: 'Hello'
- Double quotes: "Hello"
- Triple quotes: '''Hello''' (used for long paragraphs)

Important Rule: Strings are immutable. This is a fancy word that means once you create a string, you cannot change the individual letters inside it. If you want a different string, you have to create a new one!



````
#We create a variable called 'my_string' and store some text in it
my_string = "Welcome to Python"

print(my_string)  # This displays the text on the screen
````

**Accessing Characters (Indexing)**

Imagine every character in your string has a "seat number." We call this an index.

- Counting starts at 0: The first character is always at index 0.
- Counting from the back: Python is cool because you can use negative numbers! -1 is always the very last character.


```
W	   e	   l	  c	   o	  m	   e	...	n

0	   1	   2	  3	   4	  5	   6	...[6]	-1
```




````
# Let's try picking out specific letters

print("The first character is:", my_string[0])   # Index 0
print("The fourth character is:", my_string[3])  # Index 3
print("The last character is:", my_string[-1])  # Index -1

#If you try to jump to a "seat" that doesn't exist (like my_string[50]), Python will give you an IndexError.
````

#Slicing Strings

What if you want a "slice" of the string instead of just one letter? We use the syntax: string[start : end][2]

- start: Where you begin (inclusive—we keep this letter).
- end: Where you stop (exclusive—we stop just before this index).

````
# Getting a piece of the string
print("Indices 1 to 5:", my_string[1:5])   # Characters from index 1 up to (but not including) 5
print("Middle section:", my_string[5:-2])  # From index 5 to the 2nd to last letter

# Pro Tip: Omitting numbers
print("From the start:", my_string[:7])    # Start to index 7
print("Until the end:", my_string[11:])    # Index 11 to the very end
````

## 1.7 Conversion of Types

We can change the type of the object in Python; this is called typecasting. For example, we can convert an integer into a float (e.g. 2 to 2.0).

````
# Example 1
# Let's first verify that this is an integer by running this cell.
#
type(2)
````

````
float(2)
````

### 1.7a. Converting integers to floats


````
type(float(2))
````

````
# Example 1
# Let's now convert integer 2 to a float and check its type
#
type(float(2))
````

When we convert an integer into a float, we don't really change the value (i.e., the significand) of the number. However, if we cast a float into an integer, we could potentially lose some information. For example, if we cast the float 1.1 to integer we will get 1 and lose the decimal information (i.e., 0.1):

````
type(1.1)
````

````
type(int(1.1))
````

````
# Example 2
# Casting 1.1 to integer will result in loss of information
#
int(1.1)
````

### 1.7d. Boolean Casting
We can cast boolean objects to other data types. If we cast a boolean with a value of True to an integer or float we will get a one. If we cast a boolean with a value of False to an integer or float we will get a zero. Similarly, if we cast a 1 to a Boolean, you get a True. And if we cast a 0 to a Boolean we will get a False. Let's give it a try:



````
# Example 1
# Let's cast 1 to a Boolean
bool(1)
````
