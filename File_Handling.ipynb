# **File Handling in Python**

**1. What is a File?**

A file is a collection of data stored on a secondary storage device (e.g., hard drive, SSD, etc.). Files are a way to permanently save data, making it retrievable for future use.

**File Object:** In Python, a file object is an interface through which Python interacts with the file on the disk. When you use functions like open(), you create a file object, which allows reading from or writing to the actual file on disk.

**Distinction between file and file object:** A file is the actual resource (like a text file), whereas a file object is the handle or reference used by the program to interact with the file.

**2. Types of Files**

**Text Files:**

Text files consist of characters that are human-readable. They are used to store plain data that can be viewed using text editors like Notepad, VS Code, or even through Python itself.

Examples: .txt, .csv, .py (Python scripts), .html.

Text files are stored as a series of bytes that represent characters, using encoding standards like ASCII (American Standard Code for Information Interchange) or Unicode.

For example, in ASCII, the character 'A' is stored as the byte 1000001, which corresponds to 65 in decimal.

Line endings: Each line in a text file is typically terminated with a newline character (\n).

**Binary Files:**

Files containing non-text data (e.g., images, audio, videos).
Not human-readable and require specialized software to read.Binary files contain data in a non-human-readable format.

The data is stored as a series of bytes representing actual information, such as images, audio, videos, or executable files.

Examples: .jpg, .mp3, .mp4, .exe.

The data is stored in raw binary, which makes it unreadable by humans but interpretable by specific programs or software.

**File Modes**
When you open a file in Python, you specify a mode to define how you want to interact with the file.

**Mode	Description**

- 'r'	Read mode (default). Opens the file for reading.
- 'w'	Write mode. Opens the file for writing (creates a new file if it doesn't exist or overwrites the existing one).
- 'a'	Append mode. Opens the file to add new content at the end, preserving existing content.
- 'x'	Exclusive creation mode. Creates a new file, but raises an error if the file already exists.
- 'b'	Binary mode. This can be used in conjunction with other modes for binary files (e.g., 'rb', 'wb').
- 't'	Text mode (default). You don’t have to specify this mode because it's the default for text files.

Example of opening a file in read mode:


```
file = open("example.txt", "r")
```

To perform any operation on a file, first, you need to open it using Python’s open() function.

````python
file = open("/content/918098387-Medical-Letter.txt", "r")
#"filename.txt": The name (and optionally, the path) of the file.
#"r": The mode for opening the file (you can change this depending on the operation you want to perform).
````

**Working with Specific File Types**

**1. text files.**

This example shows how to:

- Write to a text file
- Read from a text file
- Append to a text file

In this case, the file will store both simple text and more structured data in a CSV-like format (comma-separated values).

````python
# 1. Writing to a text file
with open("example.txt", "w") as file:#opens file in write mode.This overides the file if it exists and opens a new file if it doesn't
    # Writing simple text data
    file.write("This is a simple text file.\n")#used to write both simple text and structured data(csv format to the file)

    # Writing structured data like CSV format
    file.write("Name,Age,City\n")
    file.write("Alice,30,New York\n")
    file.write("Bob,25,Los Angeles\n")
    file.write("Charlie,35,Chicago\n")

# 2. Reading from a text file
with open("example.txt", "r") as file:
    content = file.read()  # Read the entire content of the file
    print("File content after writing:\n")
    print(content)

# 3. Appending to a text file
with open("example.txt", "a") as file:
    # Appending a new line of structured data
    file.write("David,40,Boston\n")

# 4. Reading the updated file
with open("example.txt", "r") as file:
    updated_content = file.read()  # Read the updated content
    print("\nFile content after appending:\n")
    print(updated_content)
````

````python
import shutil
import csv

# 1. Create a text file and write student's name and age into it
with open("student_record.txt", "w") as file:
    file.write("Name: John Doe\n")
    file.write("Age: 21\n")

# 2. Read the text file and print its contents
with open("student_record.txt", "r") as file:
    content = file.read()
    print("Text File Content after writing:\n")
    print(content)

# 3. Append new information to the text file (Student's grade)
with open("student_record.txt", "a") as file:
    file.write("Grade: A\n")

# 4. Read the updated file
with open("student_record.txt", "r") as file:
    updated_content = file.read()
    print("\nUpdated Text File Content:\n")
    print(updated_content)

# 5. Create a CSV file with student data
csv_data = [
    ["Name", "Age", "Grade"],
    ["Alice", 22, "A"],
    ["Bob", 24, "B"],
    ["Charlie", 23, "A"]
]

# Write CSV data into a file
with open("students.csv", "w", newline='') as file:
    writer = csv.writer(file)
    writer.writerows(csv_data)

print("\nCSV file 'students.csv' created successfully!")

# 6. Move the CSV file to a downloadable location in Colab
shutil.move("students.csv", "/content/students.csv")

# Provide the link to download the file
print("Download the CSV file here: /content/students.csv")
````

**CSV Files**

CSV (Comma-Separated Values) files store tabular data in plain text, where each line represents a row, and values are separated by commas. Python’s pandas library is great for handling CSV data.

````python
import pandas as pd
df = pd.read_csv("/content/students.csv")
print(df)

#pd.read_csv() is a function from the pandas library used to read CSV files and return the data in a DataFrame.
#DataFrame: The output of pd.read_csv() is a pandas DataFrame, which is a 2-dimensional, labeled data structure.
#It is similar to a table in a database, or a spreadsheet, and it is very powerful for data manipulation, analysis, and visualization.

#Key Features:

#The DataFrame can store heterogeneous data (strings, integers, floats, etc.).
#It automatically handles missing values (NaN).
#It allows you to use labels for rows and columns, which makes it easy to perform operations on specific rows, columns, or subsets of data.
#Automatic Parsing: read_csv() automatically detects the types of columns (such as integer, string, float, etc.).

````

````python
import numpy as np
data = np.genfromtxt("/content/students.csv", delimiter=",")
print(data)


#np.genfromtxt() is a function from the NumPy library that can be used to read CSV files into a NumPy array.
#NumPy Array: The output of np.genfromtxt() is a NumPy array, which is a more basic data structure than a DataFrame.
#It is typically used for numerical data and supports mathematical operations, but lacks the rich features of pandas, such as handling missing values or labeled data.

#Key Features:

#np.genfromtxt() is more basic compared to pandas and does not automatically infer column types (you have to specify dtype explicitly if needed).
#It works well for numerical data, but it is not as flexible with mixed data types (e.g., strings and numbers in the same dataset).
#It requires handling missing data manually, unlike pandas which uses NaN for missing values automatically.
#It provides a raw array, not labeled data, so it's harder to do operations like filtering by column name.
````

**XML Files**

XML (Extensible Markup Language) is used to encode data in a readable format. XML is structured with tags to define data.

````python
import xml.etree.ElementTree as ET #This line imports the ElementTree module from the xml.etree package and aliases it as ET.
                                   #This is a commonly used Python library for working with XML data.
                                   #xml.etree.ElementTree: This module provides classes and methods to parse and create XML documents.
                                   #ElementTree: A class in the ElementTree module used to parse and navigate an XML document.
tree = ET.parse('data.xml') #This line parses the XML file named data.xml and creates an ElementTree object, which represents the entire XML document.
                             #ET.parse(): This function reads the XML file ('data.xml') and creates an ElementTree object.
                             #This object contains the entire structure of the XML document and allows you to manipulate and extract data from it.
root = tree.getroot() #This line retrieves the root element of the XML document, which is the top-level element that contains all other elements (children) of the XML tree.
                      #tree.getroot(): This method of the ElementTree object gets the root element of the XML document. The root is the starting point for traversing the XML structure.
print(root.tag) #This line prints the tag of the root element. A tag refers to the name of an XML element.
                 #root.tag: The tag property of the root element returns the name of the root element in the XML document.
````
