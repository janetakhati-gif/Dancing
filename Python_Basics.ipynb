# Python Basics

##  1. Python Statements

A **statement** in Python is an instruction that the interpreter can execute. For example:

```python
a = 5  # assignment statement
print(a)  # function call statement
```

Multiple statements can go on one line using a semicolon `;`, but it's not recommended.

```python
x = 10; y = 20; print(x + y)
```

##  2. Indentation

Python uses **indentation** (spaces or tabs) to define blocks of code.

Incorrect indentation leads to errors.

```python
# Correct
if True:
    print("Hello, world!")

# Incorrect (Uncomment to see error)
# if True:
# print("This will cause an IndentationError")
```

##  3. Comments

Comments are used to explain code. Python ignores comments during execution.

There are two types:
- **Single-line comments** using `#`
- **Multi-line comments** using triple quotes `'''` or `"""`

```python
# This is a single-line comment

"""
This is a
multi-line comment
"""
```

##  4. Import System

Python has a powerful **import system** to reuse modules.

There are 3 ways to import:

```python
# Import entire module
import math
print(math.sqrt(16))

# Import specific function
from math import pi
print(pi)

# Rename during import
import math as m
print(m.sin(0))
```

##  5. Input and Output

You use `input()` to get user input and `print()` to display output.

```python
# Output
print("Welcome to Python!")

# Input
name = input("Enter your name: ")
print("Hello,", name)
```

##  6. Variables – Concept and Behavior

A **variable** is a name that refers to a value. Python is dynamically typed.

You don't need to declare the type – Python does it automatically.

```python
x = 10        # integer
name = "Bob"  # string
pi = 3.14     # float

print(x, name, pi)

# Reassigning a variable to a different type
x = "Now I'm a string"
print(x)
```

##  7. Constants and Naming Variables

Python doesn't have built-in constant types, but naming convention is used:

- Use UPPERCASE for constants (by convention)
- Use lowercase_with_underscores for variables

```python
MAX_USERS = 100  # This is a constant by convention
current_user = "Felista"

print(f"Max allowed: {MAX_USERS}, Current user: {current_user}")
```

##  8. Assigning a Value to a Variable

You assign a value using the `=` operator. Python automatically determines the variable type.

```python
x = 42
name = "Alice"
print(x, name)
```

##  9. Reassigning a Value to a Variable

You can reassign a new value to the same variable name.

```python
x = 10
print("Before reassignment:", x)
x = 99
print("After reassignment:", x)
```

##  10. Assigning Multiple Values to Multiple Variables

You can assign multiple values in one line.

```python
a, b, c = 1, 2, 3
print(a, b, c)
```

##  11. Assigning the Same Value to Multiple Variables

You can assign the same value to multiple variables at once.

```python
x = y = z = 100
print(x, y, z)
```

##  12. Checking the Type of a Variable

Use the built-in `type()` function to check the data type of a variable.

```python
x = 7
name = "Data"
print(type(x))
print(type(name))
```

##  13. Arithmetic Operators

Used to perform mathematical operations:
- `+` Addition
- `-` Subtraction
- `*` Multiplication
- `/` Division
- `//` Floor Division
- `%` Modulus
- `**` Exponentiation

```python
a = 10
b = 3
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
```

##  14. Comparison Operators

Used to compare values:
- `==` Equal
- `!=` Not equal
- `>` Greater than
- `<` Less than
- `>=` Greater than or equal to
- `<=` Less than or equal to

```python
x = 5
y = 10
print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

##  15. Logical Operators

Used to combine conditional statements:
- `and` Returns True if both statements are true
- `or` Returns True if one of the statements is true
- `not` Reverses the result

```python
x = 5
y = 10
print(x > 0 and y > 0)
print(x > 0 or y < 0)
print(not(x > 0))
```

##  16. Membership Operators

Used to test if a value is in a sequence:
- `in` Returns True if the value exists in the sequence
- `not in` Returns True if the value does not exist in the sequence

```python
fruits = ["apple", "banana", "mango"]
print("banana" in fruits)
print("grape" not in fruits)
```
