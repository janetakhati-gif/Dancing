# Functions in Python


## What is a Function?

A **function** is like a small machine that takes in some input, does some work, and then gives back an output.Its a reusable block of code designed to perform a specific task.

It helps us avoid repeating the same code again and again. Instead of writing the same set of instructions multiple times, we just put them inside a function and call it whenever we need it.

For example, if you keep baking cakes using the same recipe, you don't rewrite the recipe each time—you just reuse it. A function works the same way in programming.

## Why Use Functions?
1. **Reusability** → Write once, use many times.
2. **Organization** → Break your program into smaller parts.
3. **Simplicity** → Makes your code easier to read.
4. **Debugging** → Easier to find and fix mistakes.

````
## Defining a Function
#In Python, we use the keyword `def` to create (define) a function.

# Example: A simple function

def greet():
    print("Hello, welcome to Python!")

# Calling the function
greet()

### Notes:

#`def` → tells Python we are defining a function.
#`greet` → the function name.
#`()` → parentheses (can hold inputs if needed).
#`print(...)` → the task inside the function.
#`greet()` → calling (using) the function.
````

````
## Function with Parameters (Inputs)
#Sometimes, we want to give information to the function to work with. This is done using **parameters**.

def greet_user(name):
    print("Hello, " + name + "!")

# Call the function with an input
greet_user("Alice")
greet_user("Bob")


### Notes:
#`name` is a **parameter** (like an empty box that gets filled when the function is called).
#`greet_user("Alice")` → here, `Alice` is an **argument** (the actual value).
````

````

## Function with Multiple Parameters
#We can pass more than one input.

def add_numbers(a, b):
    result = a + b
    print("The sum is:", result)

add_numbers(5, 10)
add_numbers(12, 8)

### Notes:
#`a` and `b` are two parameters.
# When calling, we must give **two values**.
````

````

## Function with Return Value
#Instead of only printing, a function can also **return a result** for later use.

def square(number):
    return number * number

# Store the result in a variable
result = square(6)
print("The square is:", result)


### Notes:
#`return` → sends back a value.
# We can store that value in a variable.
#Useful when we want to use the result somewhere else in the program.
````

````
## Default Parameters
# We can set a **default value** so if no input is given, the function still works.

def greet_user(name="Friend"):
    print("Hello, " + name + "!")

# Calling with and without input
greet_user("Alice")
greet_user()


### Notes:
#If no name is given, it uses `Friend`.
````

````
## Keyword Arguments
# We can specify arguments by their names.

def introduce(name, age):
    print("My name is", name, "and I am", age, "years old.")

# Using keyword arguments
introduce(age=20, name="John")


### Notes:
# Useful when we don’t want to follow the exact order of parameters.
````

````
## Variable Number of Arguments
#Sometimes, we don’t know how many inputs the user will provide. Python lets us handle this.

def add_many(*numbers):
    total = sum(numbers)
    print("The total is:", total)

add_many(1, 2, 3)
add_many(10, 20, 30, 40)


### Notes:
#`*numbers` means we can pass **any number of inputs**.
# Inside, they behave like a list.
````

````
## Functions Inside Functions
# Functions can call other functions.

def multiply(a, b):
    return a * b

def square_and_multiply(x):
    return multiply(x, x)

print(square_and_multiply(5))

### Notes:
#We reused `multiply()` inside another function.
````

````
## Anonymous Functions (Lambda)
# Sometimes, we need a very short function that we use only once. We can use **lambda functions**.


square = lambda x: x * x
print(square(5))


### Notes:
#`lambda` creates a small one-line function.
#Used for quick tasks
````
